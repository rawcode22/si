KNN REGRESSOR

```python
cpu = read_csv('si/datasets/cpu/cpu.csv',sep=',',features=True, label=True)
from si.src.models.knnregressor import KNNRegressor

dataset_train, dataset_test = train_test_split(cpu, test_size=0.2)

#normal
knn = KNNRegressor(k=3)
knn.fit(dataset_train)

score = knn.score(dataset_test)
print(f'RMSE: {score}')

#weighted
knn = KNNRegressor(k=3, weights='distance')
knn.fit(dataset_train)
score = knn.score(dataset_test)
print(f'RMSE (weighted): {score}')
```
